```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
x = np.linspace(-2, 2, 10)
y = x**2
y2 = x
plt.plot(x, y2)
plt.plot(x, y)
plt.title("graph of x^2 vs x")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
nbpt = int(input("nb de points : "))
x = np.array(np.linspace(-1, 1, nbpt))
y = np.cos(2*math.pi*x)
plt.plot(x, y)
plt.title("cos(2pix)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
nbpt = int(input("nb de points : "))
x = np.array(np.linspace(-1, 1, nbpt))
y1 = np.cos(2*math.pi*x)
y2 = np.sin(2*math.pi*x)

plt.plot(x, y1, label="cos(2pix)")
plt.plot(x, y2, label="sin(2pix)")

plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")

plt.show()
```

```python
try : 
    nbpt = int(input("nb de points : "))
    x = np.array(np.linspace(0, 4, nbpt))

    y1 = np.sin(math.pi*x)*np.sin(math.pi*x*20)*np.exp(-x)

    plt.plot(x, y1, label="f(x)")

    plt.legend()
    plt.xlabel("x")
    plt.ylabel("f(x)")

    plt.show()

except ValueError as error:
    print(error)
```

```python
try : 
    nbpt = int(input("Number of points : "))
    temp = float(input("Temperature : "))
    vm = np.array(np.linspace(2, 10, nbpt))
    R = 0.08206

    p = R*temp/vm

    plt.plot(vm, p, label="Isotherm (ideal gaz)")

    plt.legend()
    plt.title("isotherm of ideal gaz")
    plt.xlabel("Vm (L/mol)")
    plt.ylabel("p (Pa)")

    plt.show()
    plt.savefig("isotherm of ideal gaz")

except ValueError as error:
    print(error)
```

```python
try : 
    nbpt = int(input("Number of points : "))

    f = False
    while f != True:
        temp = float(input("Temperature : "))
        vm = np.array(np.linspace(2, 10, nbpt))
        R = 0.08206

        p = R*temp/vm

        plt.plot(vm, p, label="Isotherm %s (ideal gaz)" %temp)        
        plt.legend()
        plt.title("isotherm of ideal gaz")
        plt.xlabel("Vm (L/mol)")
        plt.ylabel("p (Pa)")

    else : 
        f = True

    plt.show()
    plt.savefig("isotherm of ideal gaz")

except ValueError as error:
    print(error)
```

```python
try : 
    nbpt = int(input("Number of points : "))
    start = int(input("start interval : "))
    end = int(input("end interval : "))   
    x0 = int(input("x0 : "))
    s = float(input("s : "))


    x = np.array(np.linspace(start, end, nbpt))

    y = ((1/math.sqrt(2*math.pi))*np.exp(-(1/2)*((x-x0)**2/s**2))) 

    plt.plot(x, y)
    plt.title("Gaussian function")
    plt.xlabel("x")
    plt.ylabel("f(x)")

    plt.show()

except ValueError as error:
    print(error)
```

```python
nbpt = int(input("nb de points : "))
x = np.array(np.linspace(0, 3, nbpt))
y1 = np.exp(-x)
y2 = np.sin(3*math.pi*x)*np.exp(-x)

plt.plot(x, y1, label="e-x")
plt.plot(x, y2, label="sin(3pix)e-x")

plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")

plt.show()
```

```python
try : 
    nbpt = int(input("Number of points : "))
    start = int(input("start interval : "))
    end = int(input("end interval : "))   

    f = False
    while f != True:
        x0 = int(input("x0 : "))
        s = float(input("s : "))


        x = np.array(np.linspace(start, end, nbpt))

        y = ((1/math.sqrt(2*math.pi))*np.exp(-(1/2)*((x-x0)**2/s**2))) 

        plt.plot(x, y, label="x0 = %s, s = %s" %(x0, s))
        plt.title("Gaussian function")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.legend()

    else : 
        f = True


    plt.show()

except ValueError as error:
    print(error)
```
